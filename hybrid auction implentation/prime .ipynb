```python
from math import isqrt
def isPrime(n: int) -> bool:
    if n <= 3:
        return n > 1
    if n % 2 == 0 or n % 3 == 0:
        return 'non-prime'
    limit = isqrt(n)
    for i in range(5, limit+1, 6):
        if n % i == 0 or n % (i+2) == 0:
            return 'non-prime'
    return 'prime'
```

```python
for i in range(3,10000):
    temp = (7*i)+(7+i)
    k = isPrime(temp)
    print(temp,i,k)
```
